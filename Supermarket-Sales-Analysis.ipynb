```python
# menghandling outlier pada kolom total dengan tukey's rule

q3 = df['Total'].quantile(0.75)
q1 = df['Total'].quantile(0.25)

iqr = q3 - q1

min_boundary = q1 - (iqr * 1.5)
max_boundary = q3 + (iqr * 1.5)

print(f'Minimum boundary kolom total adalah: {min_boundary}')
print(f'Maximum boundary kolom total adalah: {max_boundary}')

# mendefinisikan data yang outlier untuk menghitung total outlier
outlier_total = df[(df['Total'] < min_boundary) | (df['Total'] > max_boundary)]

print(f'Total outlier {len(outlier_total)} data')
print(f'Persentase outlier pada kolom total adalah sebesar {len(outlier_total)*100/len(df["Total"]):.2f}%%')
```

Pada hasil tersebut terlihat bahwa hanya terdapat 9 data outlier, namun saya akan tetap menghandling data tersebut. Berikutnya saya akan mendefinisikan dataframe baru tanpa outlier dan cek outlier kolom gross income.

```python
# mendefinisikan dataframe baru tanpa outlier
df_no_outlier = df[(df['Total'] >= min_boundary) & (df['Total'] <= max_boundary)]
```

## Faktor apa saja yang dapat menyebabkan pelanggan berbelanja dengan kuantitas â‰¥ 5?

Dalam menjawab pertanyaan ini, saya akan mengubah kolom Quantity menjadi variabel kategorik dan saya akan analisis dengan uji chi-square pada kolom dengan value kategorikal, seperti invoice ID, branch, city, customer type, gender, product line, dan payment. Sedangkan pada kolom bernilai numerik, saya akan menggunakan uji korelasi spearman.

Untuk melakukan pengujian, perlu mendefinisikan hipotesis:

$H_0$: Variabel dan `Quantity` bersifat independen

$H_1$: Variabel dan `Quantity` saling bergantung (dependen)

Pengujian menggunakan tingkat kepercayaan 95%.

```python
# mengubah nilai Quantity menjadi kategorikal
df_no_outlier['Quantity_Categoric'] = df_no_outlier['Quantity'].apply(lambda x: 1 if x >= 5 else 0)
```

Saya membagi menjadi 2 kategori agar memudahkan, dan batasan di angka 5 karena berdasarkan median quantity yaitu 5.

```python
# melihat jumlah data pada kolom Quantity
df_no_outlier['Quantity_Categoric'].value_counts()
```

```python
# melakukan pengujian kolom kategorikal dengan Quantity menggunakan uji chi-square

cols = df_no_outlier.drop(columns=['Unit price','Quantity','Total', 'Date', 'Payment', 'gross income', 'Rating', 'Quantity_Categoric']).columns

for col in cols:
  cross = pd.crosstab(df_no_outlier[col],df_no_outlier['Quantity_Categoric'])
  pval = stats.chi2_contingency(cross).pvalue
  prompt = "and they are dependent" if pval<0.05 else ""
  print(f"P-value of {col} and Quantity_Categoric: {pval} {prompt}")
```

Dari pengujian chi-square tersebut, hanya variabel gender yang saling berkaitan dengan quantity. Oleh karena itu, invoice ID, Branch, City, Customer Type, Product line, dan Rating tidak memengaruhi pelanggan dalam menentukan kuantitas berbelanja. Mungkin diperlukan observasi tambahan untuk melihat faktor-faktor yang belum terdapat dalam dataframe agar lebih banyak pengaruh yang terlibat.

Untuk mengetahui cara kerja hubungan variabel gender dan kuantitas produk yang dibeli, perlu menghitung tingkat jumlah pembelian dengan mengukur rata-rata 'Quantity'. Rata-ratanya adalah tingkat jumlah pembelian produk.

```python
# melihat perbedaan mean di setiap gender
df_no_outlier.groupby('Gender')[['Quantity']].mean()
```

```python
sns.barplot(x='Gender', y='Quantity', data=df_no_outlier, errorbar=None)
plt.title('Mean Quantity by Gender')
plt.xlabel('Gender')
plt.ylabel('Mean Quantity')
plt.show()
```

Dari hasil tersebut hanya sedikit terlihat perbedaan antara perempuan dan laki-laki. Rata-rata kuantitas berbelanja pada perempuan adalah 5,6 dan laki-laki 5,2.

Selanjutnya melihat hubungan dengan variabel numerik.

```python
# melakukan pengujian kolom numerik dengan Quantity menggunakan uji spearman

cols = df_no_outlier.drop(columns=['Invoice ID', 'Branch', 'City', 'Customer type', 'Gender', 'Product line', 'Date', 'Quantity', 'Quantity_Categoric']).columns

for col in cols:
    spearman_corr, pval = stats.spearmanr(df_no_outlier[col], df_no_outlier['Quantity'])
    prompt = "and they are dependent" if pval < 0.05 else ""
    print(f"P-value of {col} and Quantity: {pval} {prompt}")
```

Dari pengujian spearman, hanya 2 variabel yang berkaitan yaitu total dan gross income. Namun, total dan gross income bukan merupakan faktor melainkan dampak yang diakibatkan dari kuantitas belanja. Semakin banyak pelanggan menambah kuantitas belanjanya, semakin banyak pula total dan gross income.

```python
# membuat definisi untuk melakukan analisis lanjutan masing-masing variabel yang dependen
def corr(col):
    tmp = df_no_outlier.groupby(col)[['Quantity']].mean().reset_index()
    spearman_corr, pval = stats.spearmanr(tmp[col], tmp['Quantity'])
    print('Spearman Correlation:', spearman_corr)
    print('P-value:', pval)
```

```python
# melihat hubungan variabel total dan quantity
corr('Total')
```

```python
sns.scatterplot(x='Total', y='Quantity', data=df_no_outlier)
plt.title('Total vs Quantity')
plt.xlabel('Total')
plt.ylabel('Quantity')
plt.show()
```

Pada grafik terlihat bahwa semakin tinggi quantity, semakin banyak pula totalnya

```python
# melihat hubungan variabel gross income dan quantity
corr('gross income')
```

```python
sns.scatterplot(x='gross income', y='Quantity', data=df_no_outlier)
plt.title('Gross Income vs Quantity')
plt.xlabel('Gross Income')
plt.ylabel('Quantity')
plt.show()
```

Dari perhitungan tersebut, nilai korelasi spearman pada kolom total dan gross income sebesar 0,634 yang berarti menunjukkan hubungan positif yang kuat yang mengindikasikan ketika jumlah barang yang dibeli (Quantity) meningkat, maka total dan gross income juga akan meningkat begitupun sebaliknya.

P-value yang sangat kecil yaitu 9.248548737044144e-112 menunjukkan bahwa hubungan antara Total dan Quantity serta hubungan gross income dan quantity sangat signifikan secara statistik. Artinya, kemungkinan besar hubungan yang ditemukan tidak terjadi secara kebetulan.


Berdasarkan nilai-nilai korelasi tersebut, dapat diketahui bahwa jumlah pembelian produk memiliki korelasi terhadap gender, total, dan gross income.

## Mengapa kuantitas belanja para pelanggan penting bagi supermarket "EasyMart"?

```python
def corr(col):
    tmp = df_no_outlier.groupby(col)[['Quantity']].mean().reset_index()
    spearman_corr, pval = stats.spearmanr(tmp[col], tmp['Quantity'])
    print('Spearman Correlation:', spearman_corr)
    print('P-value:', pval)

corr('gross income')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df_no_outlier['Quantity'], df_no_outlier['gross income'], color='skyblue', alpha=0.6)
plt.title('Hubungan antara Kuantitas Belanja dan Pendapatan Kotor')
plt.xlabel('Kuantitas Belanja')
plt.ylabel('Pendapatan Kotor')
plt.grid(True)
plt.show()
```

Seperti perhitungan pada pertanyaan pertama, hasil tersebut menunjukkan bahwa kuantitas belanja para pelanggan cukup penting bagi supermarket "EasyMart" karena berkorelasi positif dan berhubugan secara signifikan dengan pendapatan kotor supermarket (p-value < 0,05 yaitu 1.0055745624637118e-164). Hal tersebut mengindikasikan bahwa secara signifikan apabila kuantitas produk yang dibeli pelanggan akan meningkatkan gross income bagi supermarket "EasyMart".
